# Taxa de escolarização

#### 2977 - Pessoas que frequentavam creche ou escola por nível de ensino, cor ou raça e grupos de idade
#### 200 - População residente, por sexo, situação e grupos de idade - Amostra - Características Gerais da População
##### [população residente no município de 6 a 14 anos de idade matriculada no ensino regular/total de população residente no município de 6 a 14 anos de idade] x 100

##### OBS.: [Tabela 2977 / Tabela 200] x 100  --> PARA RESULTADO EM % (como consta no IBGE), porém, para execução de cálculos futuros, o resultado está em DECIMAL.

```python
import pandas as pd
```

```python
populacao_df = pd.read_csv ('.\dados_brutos\Tabela 200.csv', sep= ';', dtype= 'str', encoding='ISO-8859-1')
```

```python
populacao_df
```

```python
lista= ['Cód.', 'População Residente']
populacao_df = populacao_df [lista]

populacao_df['População Residente'] = populacao_df['População Residente'].astype(float)
populacao_df = populacao_df.dropna()
```

```python
populacao_df.dtypes
```

```python
ensino_df = pd.read_csv ('.\dados_brutos\Tabela 2977.csv', sep= ';', dtype= 'str', encoding='ISO-8859-1')
```

```python
ensino_df
```

```python
lista1= ['Cód.', 'População']
ensino_df = ensino_df [lista1]
```

```python
ensino_df['População'] = ensino_df['População'].astype(float)
ensino_df.dropna()
```

```python
df_all= pd.merge(ensino_df, populacao_df, left_on= 'Cód.', right_on= 'Cód.', how= 'inner' )
```

```python
df_all
```

```python
df_all['schooling_rate_2010']= df_all['População']/df_all['População Residente']

df_all['schooling_rate_2010'] = round(df_all['schooling_rate_2010'],4)

df_all = df_all.rename(columns= {'Cód.':'cod_ibge'})

_list=['cod_ibge', 'schooling_rate_2010']
df_all = df_all[_list]
```

```python
df_all.to_json('taxa_escolarizacao_2010.json', orient='records')
```
